```python
import numpy as np
import pandas as pd

df = pd.read_csv('/Employee.csv')

# ใส่ df.drop() ถ้าไม่เอาข้อมูลในแถวไหน

display(df)

```

```python
print(df.info())
```

```python
df = pd.get_dummies(df , columns = ['Education', 'City', 'Gender', 'EverBenched'])
print(df)
```

```python
x = df.drop(columns = 'LeaveOrNot')
x = np.array(x)
y = df['LeaveOrNot']
y = np.array(y)

print(x,y)
```

```python
unique, counts = np.unique(y , return_counts = True)
print(unique, counts)
```

```python
from sklearn.preprocessing import MinMaxScaler

scalerx = MinMaxScaler()
scalerx.fit(x)
x = scalerx.transform(x)
print(x)
```

```python
from sklearn.model_selection import train_test_split

xtrain, xtest, ytrain, ytest = train_test_split(x, y, test_size = 0.2, random_state = 42)
print(xtrain, xtest, ytrain, ytest)
```

```python
from sklearn.neural_network import MLPClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

# Create the models
mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), activation = 'logistic', max_iter=1000, random_state=42,
                          learning_rate_init = 0.01)
svc_model = SVC(kernel='rbf', random_state = 42)
knn_model = KNeighborsClassifier(n_neighbors=5)

# Train the models
mlp_model.fit(xtrain, ytrain)
svc_model.fit(xtrain, ytrain)
knn_model.fit(xtrain, ytrain)

# Make predictions
mlp_predictions = mlp_model.predict(xtest)
svc_predictions = svc_model.predict(xtest)
knn_predictions = knn_model.predict(xtest)

# Evaluate the models
mlp_accuracy = accuracy_score(ytest, mlp_predictions)
svc_accuracy = accuracy_score(ytest, svc_predictions)
knn_accuracy = accuracy_score(ytest, knn_predictions)

print('MLP Classification')
print("MLP Accuracy:", mlp_accuracy)
print('Predicted Labels:', mlp_predictions)
print('Actual Labels:', ytest)

print('SVC Classification')
print("SVC Accuracy:", svc_accuracy)
print('Predicted Labels:', svc_predictions)
print('Actual Labels:', ytest)

print('KNN Classification')
print("KNN Accuracy:", knn_accuracy)
print('Predicted Labels:', knn_predictions)
print('Actual Labels:', ytest)






```

```python
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report

print(confusion_matrix(ytest, mlp_predictions))
print(classification_report(ytest, mlp_predictions))
```

```python
import pandas as pd # Import pandas
from sklearn.model_selection import KFold
from sklearn.model_selection import cross_val_score
from sklearn.neural_network import MLPClassifier
from sklearn.metrics import accuracy_score, make_scorer, f1_score
from sklearn.preprocessing import MinMaxScaler # Import MinMaxScaler
import numpy as np # Ensure numpy is imported

# Re-define df, x and y, and scale them within this cell to ensure they are
# available
df = pd.read_csv('/Employee.csv') # Load the dataframe
df = pd.get_dummies(df , columns = ['Education', 'City', 'Gender', 'EverBenched']) # Apply get_dummies
x = df.drop(columns = 'LeaveOrNot')
x = np.array(x)
y = df['LeaveOrNot']
y = np.array(y)

scalerx = MinMaxScaler()
scalerx.fit(x)
x = scalerx.transform(x)

mlp_model = MLPClassifier(hidden_layer_sizes=(100, 50), activation = 'logistic', max_iter=1000, random_state=42,
                          learning_rate_init = 0.01)

# Define number of folds for cross-validation
num_folds = 5

# Create a KFold object
kf = KFold(n_splits=num_folds)

# Defining the scoring metric
scoring = make_scorer(f1_score) # This line is immediately overwritten
scoring = make_scorer(accuracy_score) # Accuracy score will be used

# Perform cross-validation
mlp_scores = cross_val_score(mlp_model, x, y, cv=kf, scoring=scoring)
print(mlp_scores)

# Calculate and print mean score
mean_score = mlp_scores.mean()
print("Mean Recall Score:", mean_score) # Note: This will print the mean accuracy score, not recall.
```
