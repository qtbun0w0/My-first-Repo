NP Arrays

```python
import torch
import numpy as np

np1 = np.random.rand(3,4)

print(np1,'\n')
np1.dtype
```

Tensors

```python
tensor34 = torch.randn(3,4)

print(tensor34,'\n')
tensor34.dtype
```

```python
tensor234 = torch.zeros(2,3,4)

tensor234
```

```python
mytensor = torch.tensor(np1)

mytensor
```
